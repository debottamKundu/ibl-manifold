```python
location = (
    "../../custom-rnn/decision_rnn/models/flowing-bush-151/checkpoints/checkpoint_epoch_39500.pt"
)
```

```python
checkpoint, model = load_checkpoint_and_model(location)
```

```python
# modify checkpoint to add noise
checkpoint["config"]["environment"]["engagement_noise"] = 1.73
```

```python
# generate a bunch

sessions = []

for idx in range(100):

    session = run_single_session(checkpoint, model)
    session = adjust_rnn_indices(session)
    session = add_bayes_prior(session)
    session = add_congruency_incongruency(session)
    session = collapse_trials_quiscience(session)

    sessions.append(session)

all_data = merge_sessions_with_quiscience(sessions)
```
