```python
import pickle as pkl
import numpy as np
from matplotlib import pyplot as plt
```

```python
engagement_og = pkl.load(open("../data/generated/all_eids_engagement.pkl", "rb"))
```

```python
engagement_new = pkl.load(
    open(
        "../../public_hybrid_rnn/session_data/extracted_motivation_scalars_predictions_check.p",
        "rb",
    )
)
```

```python
engagement_new_eids = list(engagement_new.keys())
```

```python
motivations = []
for k in engagement_new_eids:
    if k in engagement_og.keys():
        motivations.append(
            [np.asarray(engagement_new[k]["motivation_scalar"]), np.asarray(engagement_og[k])]
        )
```

```python
for idx in range(len(motivations)):
    new, og = motivations[idx]
    plt.plot(np.arange(len(new)), new)
    plt.plot(np.arange(len(og)), og)
    plt.show()
```
