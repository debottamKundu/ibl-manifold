```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edges_from(
    [
        ("Engagement", "Prior"),
        ("Prior", "Manifold"),
        ("Engagement", "Manifold"),
        ("Prior", "Decision"),
        ("Stimulus", "Decision"),
        ("Manifold", "Decision"),
    ]
)
```

```python
pos = nx.spring_layout(G)
fig, ax = plt.subplots(figsize=(8, 8))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=5000,
    font_size=10,
    arrows=True,
)
plt.show()
```

```python
for node in ["Engagement", "Prior"]:
    sep = nx.is_d_separator(G, {"Decision"}, {"Manifold"}, {node})
    print(f"Is it safe to condition on {node}? {sep}")
```
